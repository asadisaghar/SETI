```python
import random
import cProfile
from math import *
import numpy as np
import tools
import datetime
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# ===================== #
#        CONSTANTS      #
# ===================== #
pc = 1.
sec = 1.
M_solar = 1.
# Multiply everything by relevant units below to translate them into pc, s, and M_solar
km2m = 1.e3
m2km = 1./km2m
kpc2pc = 1.e3
pc2kpc = 1./kpc2pc
pc2m = 3.08567758e16
m2pc = 1./pc2m
pc2km = pc2m*m2km
km2pc = km2m*m2pc
Msolar2kg = 1.9891e30
kg2Msolar = 1./Msolar2kg
yr2sec = pi*1e7
sec2yr = 1./yr2sec
Myr2yr = 1.e6
yr2Myr = 1./Myr2yr
Myr2sec = Myr2yr*yr2sec
sec2Myr = 1./Myr2sec

Gconst = 6.67e-11 #N.m^2.kg^-2 = kg^1.m^1.s^-2.m^2.kg^-2 = kg^-1.m^3.s^-2 
cSpeed = 3.e5 #[km/s]
RS = 11
random.seed(RS)
```

```python
# ===================== #
#        FUNCTIONS      #
# ===================== #
def init_pos(N, low, high_scale, dist):
    random.seed(RS)
    if dist == 'exp':
        init = np.random.exponential(high_scale, N)
    elif dist == 'uni':
        init = np.random.uniform(low, high_scale, N)

    return init

def r_bulge_oscillation(galaxy, t, v, N_bulge, amp):
    r = galaxy[0,:N_bulge]
    omega = 2.*pi/(25.*pi*1e7*1e6)
    phase = init_pos(N_bulge, 0., 2.*pi, 'uni')
    galaxy[0,:N_bulge] = amp*np.cos(omega*t+phase)
    return galaxy[0,:N_bulge]
```

```python
# BULGE
N_bulge = 1e4
R_bulge = 30.e3  # bulge radius #[pc]
I_0_bulge = 5.e9  # bulge central intensity
alpha_bulge = R_bulge/3.  # bulge scale length #[pc]
n_s_bulge = 5  # Bulge Sersic index(in range: 1.5-10)
M_bulge = 2.e10  # bulge mass #[M_solar]
mean_bulge = 2000. # (My arbitrary value!!) #FIXME value  #[km/s]
sigma_bulge = 30.
rho_bulge = M_bulge/(4./3.*pi*R_bulge**3)

# POS:Bulge (Spherical)
r = abs(np.random.normal(0, R_bulge, N_bulge))  #galaxy[0]  #[pc]
theta = tools.init_pos(N_bulge, 0., np.nextafter(pi,4), 'uni')  #galaxy[1]  #[pc]
phi = tools.init_pos(N_bulge, 0., 2.*pi, 'uni')  #galaxy[2]  #[pc]

# VEL:Bulge (Gaussian dist. with given mean and dispersion)
Vr = np.zeros(N_bulge) #[km/s]
Vtheta = np.zeros(N_bulge) #[km/s]
Vphi = np.zeros(N_bulge) #[km/s]

galaxy = np.zeros((6, N_bulge))
galaxy[0] = r
galaxy[1] = theta
galaxy[2] = phi
galaxy[3] = Vtheta
galaxy[4] = Vr
galaxy[5] = Vphi
```

```python
t = 0
i = 0
t_f = 200.*Myr2sec
dt_r = .2*Myr2sec
x_part = []
y_part = []
z_part = []
ind = 80
print r[ind]*1e-3
lmax = R_bulge
lmin = -lmax
print lmax
while t < t_f:
    t = i*dt_r
    time = t/Myr2sec
# Evaluate bulge velocities (Spherical)
    galaxy[3] = np.ones(N_bulge)*200.
    galaxy[1] += galaxy[3]*km2pc*dt_r/galaxy[0]
#    galaxy[2] += galaxy[3]/1.04*km2pc*dt_r/galaxy[0]
    
#    galaxy[0] = r_bulge_oscillation(galaxy, t, sigma_bulge, N_bulge, R_bulge/20.)   
    r, theta, phi = galaxy[0], galaxy[1], galaxy[2]
    x = r*np.sin(theta)*np.cos(phi)
    y = r*np.sin(theta)*np.sin(phi)
    z = r*np.cos(theta)
    x_part.append(x[ind])
    y_part.append(y[ind])
    z_part.append(z[ind])
    i += 1

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_part, y_part, z_part, marker='o')
ax.set_xlim([lmin, lmax])
ax.set_ylim([lmin, lmax])
ax.set_zlim([lmin, lmax])

plt.show()
```
